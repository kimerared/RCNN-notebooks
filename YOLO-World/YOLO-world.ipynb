## Libraries

```python
from PIL import Image
from ultralytics import YOLO
import matplotlib.pyplot as plt
import glob
```

## Load Model

```python
img_path = '/content/drive/MyDrive/Group-8-Deep-Learning/TEST/helicopter-military-unsplash.jpg'
```

```python
# Initialize a YOLO-World model
model = YOLO('yolov8s-world.pt')  # or choose yolov8m/l-world.pt

# Define custom classes
model.set_classes(["person", "helicopter"])

# Execute prediction for specified categories on an image
results = model.predict(img_path, save=True)

# Show results
results[0].show()
```

```python
# Get a list of all image paths
single_image_path = glob.glob('/content/runs/detect/predict/*.jpg')


# Iterate over each image path
for image_path in single_image_path:
    # Open the image
    image = Image.open(image_path)

    # Display the image
    plt.imshow(image)
    plt.axis('off')  # Turn off axis
    plt.show()
```

## Second image

```python
source_img_path = '/content/drive/MyDrive/Group-8-Deep-Learning/TEST/red-yellow-soccer-unsplash.jpg'
```

```python
# Initialize a YOLO-World model
model = YOLO('yolov8s-world.pt')  # or choose yolov8m/l-world.pt

# Define custom classes
model.set_classes(["person in red", "ball"])

# Execute prediction for specified categories on an image
results = model.predict(source_img_path, save=True, conf=0.02)

# Show results
results[0].show()
```
